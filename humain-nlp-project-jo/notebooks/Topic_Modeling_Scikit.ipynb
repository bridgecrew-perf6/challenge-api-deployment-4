```python
import pandas as pd 
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
from sklearn import decomposition
import numpy as np
import regex as re 
import json
import nltk
import matplotlib.pyplot as plt
from nltk.stem.porter import PorterStemmer
from sklearn.model_selection import train_test_split
```

```python
nltk.download('punkt')
```

```python
#seting the maximum column width to display all columns
pd.set_option('display.max_colwidth',-1)
```

```python
#loads json file
jsn=json.loads(open('data/case_data.json').read())
#converts json to dataframe
data=pd.DataFrame.from_dict(pd.json_normalize(jsn,'data'))
#display first 4 rows
data.head(4)
```

```python
#display the rows and columns
data.shape
```

```python
#display count of description
data['description'].value_counts()
```

```python
#display the count of each function
data['function'].value_counts()
```

```python
#select independent features
new_data=data[['description','name','industries','function']]
```

```python
#distribute the features into 60-40 ratio as train and hold
X_train,X_hold=train_test_split(new_data,test_size=0.6,random_state=50)
```

```python
X_train['name'].value_counts()
```

```python
#create stemmer instance
stemmer=PorterStemmer()
```

```python
def tokenize(text):
    tokens=[word for word in nltk.word_tokenize(text) if len(word)>3]
    return tokens
```

```python
#create tfidf vectorizer class to find the fequency of the words
vectorizer_tf=TfidfVectorizer(tokenizer=tokenize,stop_words='english',max_features=150,use_idf=False,norm=None)
tf_vectors=vectorizer_tf.fit_transform(X_train.description)
```

```python
#count vectorizer
tf_vectors.A
```

```python
#get feature names
vectorizer_tf.get_feature_names()
```

```python
#build LDA model
lda=decomposition.LatentDirichletAllocation(n_components=6,max_iter=3,learning_method='online',learning_offset=50,n_jobs=-1,random_state=50)
lda_fit=lda.fit_transform(tf_vectors)
compo=lda.components_
```

```python
num_words=15
vocab=np.array(vectorizer_tf.get_feature_names())
top_words=lambda t:[vocab[i] for i in np.argsort(t)[:num_words-1:-1]]
topic_words=([top_words(t) for t in compo])
topics=[' '.join(t) for t in topic_words]
```

```python
#print the words of each significant topic
from simple_colors import *

for i,topic in enumerate(topics):
    print(blue('\033[1mTopic\033[0;0m','bold'),i)
    
    print(topic)
```

```python
colnames=["Topic" +str(i) for i in range(lda.n_components)]
docnames=["Document" + str(i) for i in range(len(X_train.description))]
doc_topic=pd.DataFrame(np.round(lda_fit,2),columns=colnames,index=docnames)
significant_topic=np.argmax(doc_topic.values,axis=1)
doc_topic['dominant_topic']=significant_topic
```
